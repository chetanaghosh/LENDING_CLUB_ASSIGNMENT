# CORRECTING DATA TYPES.


Categorical variables - revol_util, term, mths_since_last_delinq, mths_since_last_record, last_pymnt_d, last_credit_pull_d, application_type, loan_status, grade, sub_grade, emp_length, term, home_ownership, verification_status, pymnt_plan, purpose.

Among the numerical variables, int_rate, is object data type (should be int or float). All other data types are correct.

```python
df['int_rate']
```

```python
#Removing%. Converting to float64.
df['int_rate'] = df['int_rate'].str.replace('%', '')
df.int_rate = df.int_rate.astype('float64')
```

```python
#Removing outliers data which has less contribution for Data Analysis

fig, axes = py.subplots(1, 2, figsize=(10, 5))
sns.boxplot(x=df['int_rate'], data=df, ax=axes[0])
sns.boxplot(x=df['annual_inc'], data=df, ax=axes[1])
axes[1].set_xscale('log')
```

```python
df=df[df['int_rate']<23.4]
df=df[df['annual_inc']<700000]
```

```python
# As the loan_status of "fully Paid" and "Charged Off" will only be used for analysis we will be deleting the data Loan_status with current
df=df[(df['loan_status']=='Fully Paid') | (df['loan_status']=='Charged Off')]
```

```python
# Changing Employment length column according to our requirement
df['emp_length']= df['emp_length'].replace('< 1 year', 0)
df['emp_length']= df['emp_length'].replace('1 year', 1)
df['emp_length']= df['emp_length'].replace('2 years', 2)
df['emp_length']= df['emp_length'].replace('3 years', 3)
df['emp_length']= df['emp_length'].replace('4 years', 4)
df['emp_length']= df['emp_length'].replace('5 years', 5)
df['emp_length']= df['emp_length'].replace('6 years', 6)
df['emp_length']= df['emp_length'].replace('7 years', 7)
df['emp_length']= df['emp_length'].replace('8 years', 8)
df['emp_length']= df['emp_length'].replace('9 years', 9)
df['emp_length']= df['emp_length'].replace('10+ years', 10)
#
```

Using bivariate and multivariate analysis to answer the most important business question - Which people are likely to pay off the loan completely? 

### Bi Variate Analysis between Categorical Variables

```python
df_fullypaid = df[(df['loan_status'] == 'Fully Paid')]
df_chargedoff = df[(df['loan_status'] == 'Charged Off')]
```

```python
sns.countplot(x=df['grade'],data=df, hue=df['loan_status'])
```

```python
# Relative percentage of charged of loans with respect to grade
# Dividing the number of charged loans per category to number loans of that category multiplied by 100 to get relative percentage of charged loans
per_chargedoff_A=100*(df_chargedoff['grade'].value_counts().get('A',0)/df['grade'].value_counts().get('A',0))
per_chargedoff_B=100*(df_chargedoff['grade'].value_counts().get('B',0)/df['grade'].value_counts().get('B',0))
per_chargedoff_C=100*(df_chargedoff['grade'].value_counts().get('C',0)/df['grade'].value_counts().get('C',0))
per_chargedoff_D=100*(df_chargedoff['grade'].value_counts().get('D',0)/df['grade'].value_counts().get('D',0))
per_chargedoff_E=100*(df_chargedoff['grade'].value_counts().get('E',0)/df['grade'].value_counts().get('E',0))
per_chargedoff_F=100*(df_chargedoff['grade'].value_counts().get('F',0)/df['grade'].value_counts().get('F',0))
per_chargedoff_G=100*(df_chargedoff['grade'].value_counts().get('G',0)/df['grade'].value_counts().get('G',0))

print(f'Relative percentage of A grade Charged off loans are: {per_chargedoff_A}')
print(f'Relative percentage of B grade Charged off loans are: {per_chargedoff_B}')
print(f'Relative percentage of C grade Charged off loans are: {per_chargedoff_C}')
print(f'Relative percentage of D grade Charged off loans are: {per_chargedoff_D}')
print(f'Relative percentage of E grade Charged off loans are: {per_chargedoff_E}')
print(f'Relative percentage of F grade Charged off loans are: {per_chargedoff_F}')
print(f'Relative percentage of G grade Charged off loans are: {per_chargedoff_G}')
```

###### From the above analysis it is evident that as the Grade of the loan increases the percentage of charged off loan increases
